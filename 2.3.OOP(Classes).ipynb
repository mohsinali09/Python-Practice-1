<h1 align=center > OOP (CLASSES) </h1>

<b>Python</b> is an <b>“object-oriented programming language”</b>. This means that almost all the code is implemented using a special construct called <b>classes</b>.
 Programmers use <b>classes</b> to keep related things together. This is done using the keyword <b>“class”</b> which is a grouping of object-oriented constructs.

## What is a Class?
-	A Class is a model
-	A Class is a blueprint (map) of anything
-	A Class is a template
-	A Class may also be defined as "something that can be followed to create objects and instances"

## A complete example on classes & objects

### Class

```python
class Car():
    def __init__(self, make, model, year, battery):
        self.make = make
        self.model = model
        self.year = year
        self.battery = battery
    def descriptionCar(self):
        print(f"The make of this car is {self.make}")
        print(f"The model of this car is {self.model}")
        print(f"The manufacturing year of this car is {self.year}")
        print(f"The battery ampere of this car is {self.battery}")
    def move(self):
        print(f"{self.make} is moving with speed")
    def applyBreak(self):
        print(f"{self.model} has applied the break")
    def Battery(self):
        print(f"The Battery of the car is {self.battery}")
    def setBatterySize(self, newSize):
        self.battery = newSize
    def getBatterySize(self):
        print(f"The size of your car's battery is {self.battery}")
```

### Objects

```python
car1 = Car("Honda", "Civic", 2019, "200amp" )
car2 = Car("Suzuki", "Cultus", 2015, "300amp" )
```
